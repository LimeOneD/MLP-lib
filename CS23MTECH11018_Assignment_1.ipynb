```python
import random
import numpy as np
class Perceptron:
    def __init__(self, num_epochs=85):
        self.num_epochs = num_epochs
        self.weights = None
        self.bias = None

    def train(self, X, y):
        num_samples = len(X) 
        num_features = len(X[0])
        self.weights = np.array([random.random() for _ in range(num_features)])
        self.bias = 0

        for _ in range(self.num_epochs):
            for i in range(num_samples):
                if self.predict(X[i]) != y[i]:
                    update = (y[i] - self.predict(X[i]))
                    if(update > 0):
                        self.weights = self.weights + np.array(update * X[i])
                        self.bias += update

    def predict(self, x):
        activation = np.dot(self.weights, x) + self.bias
        return 1 if activation >= 0 else 0
```

```python
import random
import numpy as np
class GD:
    def __init__(self,learning_rate, num_epochs=1000):
        self.num_epochs = num_epochs
        self.learning_rate = learning_rate
        self.weights = None
        self.bias = None
    def sigmoid(self, x):
        if(x>=0):
            return 1 / (1 + np.exp(-x))
        else:
            return np.exp(x) / (1 + np.exp(x))
    
    def train(self, X, y):
        num_features = len(X[0])
        self.weights = np.array([random.random() for _ in range(num_features)])
        self.bias = 0
        
        for epoch in range(self.num_epochs):
            loss = 0;
            ep = 0.0000001
            for i in range(len(X)):
                output = self.sigmoid((X[i][0]*self.weights[0]+ X[i][1]*self.weights[1]) + self.bias)
                loss += -(y[i]*np.log(output+ep) + (1-y[i])*np.log(1-output+ep))

                dy = -(y[i]/(output+ep) - (1-y[i])/(1-output+ep))
                da = (output)*(1-output)*dy
    
                d_weights0 = X[i][0] * da
                d_weights1 = X[i][1] * da
                d_bias = da
    
                self.weights[0] -= self.learning_rate * d_weights0
                self.weights[1] -= self.learning_rate * d_weights1
                self.bias -= self.learning_rate * d_bias
            print("loss in epoch", epoch, "is", loss/len(X))

    
```

```python
#data creation
import random
data = []
for i in range(-100,100):
    for j in range(-100, 100):
        if((i**2+j**2) <=1000):
            data.append([[i,j],0])
        if((i**2 + j**2) > 5000 and (i**2+j**2)<10000):
            data.append([[i,j],1])


random.shuffle(data)
X_data = []
y_data = []

for i in range(len(data)):
    X_data.append(data[i][0])
    y_data.append(data[i][1])
X= X_data[: int(len(X_data)/10)]
y= y_data[: int(len(y_data)/10)]

import matplotlib.pyplot as plt
class_0 = [X[i] for i in range(len(X)) if y[i] == 0]
class_1 = [X[i] for i in range(len(X)) if y[i] == 1]

# Extract x and y coordinates for each class
class_0_x = [x[0] for x in class_0]
class_0_y = [x[1] for x in class_0]
class_1_x = [x[0] for x in class_1]
class_1_y = [x[1] for x in class_1]

# Plotting
plt.scatter(class_0_x, class_0_y, color='blue', label='Class 0')
plt.scatter(class_1_x, class_1_y, color='red', label='Class 1')

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Scatter plot of two classes with dataset 1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#MLP for only one hidden layer
import numpy as np
import random
import math


class Binary_classificatoin:
    def __init__(self, layers):
        self.layers = layers
        self.weights =[]
        self.bias = []
        self.a=[]
        self.z=[]
        self.res = []
        self.loss = []
        
        for i in range(len(self.layers)-1):
            weights_ = np.random.uniform(size=(self.layers[i+1], self.layers[i]))
            bias_ = np.random.uniform(size=(self.layers[i+1]))
            self.weights.append(weights_)
            self.bias.append(bias_)

    def sigmoid(self, l):
        answers = []
        for value in l:
            if value >= 0:
                sig = 1 / (1 + np.exp(-value))
            else:
                sig = np.exp(value) / (1 + np.exp(value))
            answers.append(sig)
        return answers


    def binary_crossentropy_loss(self, y_pred, y):
        epsilon = 1e-15
        return -(y*math.log(y_pred[0]+epsilon)+ (1-y)*math.log(1-y_pred[0]+epsilon))

    def crosentropy_gad(self, y_pred, y):
        return (( (1-y)/(1-y_pred) )-( y/y_pred ))

    def sigmoid_grad(self, l):
        answers = []
        for value in l:
            sig_x = self.sigmoid(value)
            sig_g = sig_x[0] * (1-sig_x[0])
            answers.append(sig_g)
        return answers

    def forward(self, X, y):
        a_ =[]
        
        for i in range(len(X)):
            a_ = X[i]
            z_temp = []
            a_temp = []
            for j in range(len(self.layers)-1):
                z_= np.dot(np.array(self.weights[j]), a_)+self.bias[j]
                z_temp.append(z_)
                a_= self.sigmoid(z_)
                a_temp.append(a_)
                if(j==len(self.layers)-2):
                    self.res.append(a_)
                    self.loss.append(self.binary_crossentropy_loss(a_, y[i]))
            self.a.append(a_temp)
            self.z.append(z_temp)
        #self.res = np.array(self.res).flatten()
            
            
    def backward(self, X, y, res, learning_rate):
        for i in range(len(y)):
            da = []
            dz = []
            dw = []
            db = []
            i_ =[]
            dal = self.crosentropy_gad(res[i][0], y[i])
            dzl = res[i][0]*(1-res[i][0])*dal
            da.append(dal)
            dz.append(dzl)
            dasl = self.weights[len(self.layers)-2] * dzl
            dzsl = self.sigmoid_grad(dasl)
            da.append(dasl)
            dz.append(dzsl)
            for j in range(len(self.layers)-3):
                da_ = np.dot(self.weights[len(self.layers)-3-j].T, np.array(dz[-1])).T
                dz_ = self.sigmoid_grad(da_)
                da.append(da_)
                dz.append(dz_)
                i_.append(j)
            
            da = da[::-1]
            dz = dz[::-1]

            for j in range(len(self.layers)-1):
                dw.append((self.weights[j].T * da[j]).T)
                db.append(da[j].flatten())
                
            for j in range(len(self.layers)-1):
                self.weights[j] -= learning_rate* dw[j]
                self.bias[j] -= learning_rate*db[j]
                   
    def fit(self, X, y, learning_rate, epoch):
        for _ in range(epoch-1):
            self.forward(X, y)
            self.backward(X,y, self.res, learning_rate)

        self.forward(X,y)


    def predict(self, X):
        y =[]      
        for i in range(len(X)):
            y_pred = X[i]
            for j in range(len(self.weights)):
                a_= np.dot(self.weights[j], y_pred) + self.bias[j]
                y_pred = self.sigmoid(a_)
            y.append(y_pred)
              
        return y         
        
```

```python
n = Binary_classificatoin([2, 4 ,1]) ## the Binary_classifier object take a list with 3 values [no_neurons_in_1st_layere, no_of_neeurons_in_2nd_lyer, no_of_neurons_in_output_layer}
                                                                                                
n.fit(X, y, .01, 100)                ## The fit method takes the following inputs (X_train, y_train, learning_rate, epochs)
pred = n.predict(X)                  ##the predict method takes input (X_test) and returns a list of predicted values
```
